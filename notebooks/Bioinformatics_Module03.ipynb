####**TASK P18.2** <font color= orange>**Complete Rosalind Problem #5  Dictionaries**</font>

````python
with open("rosalind_ini6.txt","r") as data:
  line=data.read()

counts = {} # Create an empty dictionary to store word counts

for word in line.split():
    # If the word is already in the dictionary, increment its count
    if word in counts:
        counts[word] += 1
    # If the word is not in the dictionary, add it with a count of 1
    else:
        counts[word] = 1

# Print the words and their counts
for key, value in counts.items():
      print(key,value)
````

#<font color=blue>**P23.**</font> SeqUtils.GC( )  a function to compute GC content

````python
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction

def main(filename: str):
    # Find the record with the maximum GC content using a generator expression and max()
    max_gc_record = max(SeqIO.parse(filename, "fasta"), key=lambda record: gc_fraction(record.seq))

    # Calculate the GC percentage
    max_gc = gc_fraction(max_gc_record.seq) * 100

    # Print results
    print(max_gc_record.id)
    print(f"{max_gc:.6f}")


if __name__ == "__main__":
    # Change "rosalind_gc.txt" to your dataset file
    main("rosalind_gc.txt")
````
